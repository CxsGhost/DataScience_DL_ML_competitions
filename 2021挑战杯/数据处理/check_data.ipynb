```python
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
img = keras.preprocessing.image.load_img('E:/py//2021//w0003_deal/original/W0003_0001.png')
print(img)
img = keras.preprocessing.image.img_to_array(img)

layer = keras.layers.experimental.preprocessing.Resizing(416, 416)
img = layer(img)
keras.preprocessing.image.save_img(".png", img)
print(img.shape)
print(img)


img = keras.backend.cast(img, 'int32')
plt.imshow(img)
plt.show()
```
