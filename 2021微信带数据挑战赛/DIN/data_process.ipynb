```python
feed_embed = pd.read_csv("../data/data/feed_embeddings.csv", header=0, index_col=None)
print(feed_embed)
```

```python
feed_value = feed_embed.values
print(feed_value)
print(feed_value.shape)
print(type(feed_value[0][0]))
print(type(feed_value[0][1]))
```

```python
my_feed_info = pd.read_csv("my_data/feed_info.csv", header=0, index_col=0)
my_info = pd.merge(my_feed_info, feed_embed, on=['feedid'])
print(my_info)
```

```python
print(my_feed_info['videoplayseconds'].describe())
```

```python
for i in range(my_info.shape[0]):
    temp = my_info['videoplayseconds'][i]
    if 0 < temp <= 10:
        my_info['videoplayseconds'][i] = 1
    elif 10 < temp <= 20:
        my_info['videoplayseconds'][i] = 2
    elif 20 < temp <= 30:
        my_info['videoplayseconds'][i] = 2
    elif 30 < temp <= 40:
        my_info['videoplayseconds'][i] = 4
    elif 40 < temp <= 50:
        my_info['videoplayseconds'][i] = 5
    elif 50 < temp <= 62:
        my_info['videoplayseconds'][i] = 6
    else:
        my_info['videoplayseconds'][i] = 7

print(my_info)
```

```python
my_info.to_csv("my_data/my_info.csv")
```

```python
action_data = pd.read_csv("../data/data/user_action.csv", header=0, index_col=None)
my_action = ['userid', 'feedid', 'date_', 'device', 'read_comment', 'like', 'click_avatar', 'forward']
action_data = action_data[my_action]

user_group = action_data.groupby(by=['userid'])
agg_method = {'like': 'sum', 'read_comment': 'sum', 'comment': 'sum', 'click_avatar': 'sum'}
temp = user_group.agg(agg_method)
temp.reset_index(inplace=True)

temp = temp.values
pos_id = temp[np.sum(temp[:, 1:4], axis=1) != 0][:, 0]

pos_data = action_data[action_data['userid'].isin(pos_id)]
print(pos_data.shape)
pos_data.to_csv("my_data/pos_user_data.csv", index=None)
```

```python
import pandas as pd

feed_info = pd.read_csv("../data/data/feed_info.csv", header=0, index_col=None)
```

```python
my_feed_info = pd.read_csv("my_data/my_info.csv", header=0, index_col=None)
```

```python
tag_list = my_feed_info['manual_tag_list']
```

```python
temp = feed_info[['feedid', 'machine_tag_list']][my_feed_info['manual_tag_list'] == '0']
lizi = temp[temp['feedid'] == 3286]
```

```python
temp = temp.fillna('-1')
def func(tag):
    if tag == '-1':
        return tag
    tag = tag.split(';')
    tag_ = []
    temp = []
    max_ = 0
    for t in tag:

        t = list(map(float, t.split(' ')))
        if t[1] > max_:
            max_ = t[1]
        temp = temp + t
        if t[1] >= 0.6:
            tag_.append(str(int(t[0])))
    if not tag_:
        tag_.append(str(int(temp[temp.index(max_) - 1])))
    return ' '.join(tag_)

temp['machine_tag_list'] = temp['machine_tag_list'].apply(func)
```

```python
temp = temp.values
for d in temp:
    print(d)
    my_feed_info.loc[my_feed_info['feedid'] == d[0], ['manual_tag_list']] = d[1]
```

```python
my_feed_info.to_csv("my_data/my_info.csv", index=None)
```

```python
my_info['manual_tag_list'] = my_info['manual_tag_list'].apply(lambda x: x.replace(';', ' '))
```
